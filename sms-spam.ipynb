```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("English")
def transform_text(text):
    usefull_words = []
    text = text.lower()
    words = nltk.word_tokenize(text)
    for word in words:
        if word not in stopwords:
            usefull_words.append(word)
    print(usefull_words)
    return " ".join(usefull_words)
```

```python
transform_text("I am bcgu jit kchouhan")
```

```python
df['processed'] = df['input'].apply(transform_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
X = cv.fit_transform(df['processed']).toarray()
print(X.shape)
```

```python
y = df['result'].values
y
```

```python
from sklearn.model_selection import train_test_split

# Assuming X and y are already defined
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=49)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB

# Initialize Naive Bayes classifiers
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
gnb.fit(X_train, y_train)
y_pred1 = gnb.predict(X_test)
print(accuracy_score(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
print(precision_score(y_test, y_pred1))
```

```python
mnb.fit(X_train, y_train)
y_pred2 = mnb.predict(X_test)
print(accuracy_score(y_test, y_pred2))
print(confusion_matrix(y_test, y_pred2))
print(precision_score(y_test, y_pred2))
```

```python
bnb.fit(X_train, y_train)
y_pred3 = bnb.predict(X_test)
print(accuracy_score(y_test, y_pred3))
print(confusion_matrix(y_test, y_pred3))
print(precision_score(y_test, y_pred3))
```
